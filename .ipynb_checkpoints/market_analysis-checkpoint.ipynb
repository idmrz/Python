```python
# Tarih stununu datetime formatna evir
df['Date'] = pd.to_datetime(df['Date'])

# Tarih bazl toplam satlar gruplama
sales_by_date = df.groupby('Date')['Total'].sum()

# Zaman serisi grafii
sales_by_date.plot(figsize=(10,6), title='Zamana Gre Satlar')
plt.ylabel('Toplam Sat')
plt.show()
```

```python
# Ortalama satlar virglden sonra iki basamakla gstermek iin
avg_sales = df.groupby(['Gender', 'Product line'])['Total'].mean().unstack()

# Heatmap'i olutururken annot parametresinde format belirleyelim
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(avg_sales, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Average Sales by Gender and Product Line')
plt.xlabel('Product Line')
plt.ylabel('Gender')
plt.show()
```

```python
# deme yntemine gre satlarn toplam
sales_by_payment = df.groupby('Payment')['Total'].sum()

# Pie chart ile grselletirme
sales_by_payment.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6))
plt.title('deme Yntemlerine Gre Sat Dalm')
plt.ylabel('')
plt.show()
```

```python
# Mteri tipine gre toplam satlar
sales_by_customer_type = df.groupby('Customer type')['Total'].sum()

# Bar grafii
sales_by_customer_type.plot(kind='bar', color='skyblue')
plt.title('Mteri Tipine Gre Satlar')
plt.ylabel('Toplam Sat')
plt.show()
```

```python
# Scatter plot ile brt gelir ve toplam sat arasndaki iliki
sns.scatterplot(x='Total', y='gross income', data=df)
plt.title('Brt Gelir ve Toplam Sat Arasndaki liki')
plt.show()
```
