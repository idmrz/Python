```python
# Tarih sütununu datetime formatına çevir
df['Date'] = pd.to_datetime(df['Date'])

# Tarih bazlı toplam satışları gruplama
sales_by_date = df.groupby('Date')['Total'].sum()

# Zaman serisi grafiği
sales_by_date.plot(figsize=(10,6), title='Zamana Göre Satışlar')
plt.ylabel('Toplam Satış')
plt.show()
```

```python
# Ortalama satışları virgülden sonra iki basamakla göstermek için
avg_sales = df.groupby(['Gender', 'Product line'])['Total'].mean().unstack()

# Heatmap'i oluştururken annot parametresinde format belirleyelim
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(avg_sales, annot=True, fmt=".2f", cmap="YlGnBu")
plt.title('Average Sales by Gender and Product Line')
plt.xlabel('Product Line')
plt.ylabel('Gender')
plt.show()
```

```python
# Ödeme yöntemine göre satışların toplamı
sales_by_payment = df.groupby('Payment')['Total'].sum()

# Pie chart ile görselleştirme
sales_by_payment.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6,6))
plt.title('Ödeme Yöntemlerine Göre Satış Dağılımı')
plt.ylabel('')
plt.show()
```

```python
# Müşteri tipine göre toplam satışlar
sales_by_customer_type = df.groupby('Customer type')['Total'].sum()

# Bar grafiği
sales_by_customer_type.plot(kind='bar', color='skyblue')
plt.title('Müşteri Tipine Göre Satışlar')
plt.ylabel('Toplam Satış')
plt.show()
```

```python
# Scatter plot ile brüt gelir ve toplam satış arasındaki ilişki
sns.scatterplot(x='Total', y='gross income', data=df)
plt.title('Brüt Gelir ve Toplam Satış Arasındaki İlişki')
plt.show()
```
